```
import pandas as pd
import seaborn as sns
```

```
# Importando a base de dados da tabela excel
vendas = pd.read_excel('/content/varejo.xlsx')
```

```
# Chamando a tabela
vendas.head()
```

```
# Chamando as informaes da tabela para analizar os valores nulos e para ver os dados de cada coluna. Tipo: str,object,float,int.
vendas.info()
```

```
# Agrupando Idcanalvenda com idcompra.
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```
# Modificando a abreviao de APP para todos os dados ficar iguais como Aplicativo
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```
# Agrupando as bandeira com idcompra
vendas.groupby("bandeira").idcompra.nunique()
```

```
# Agrupando Nome_Departamento com idcompra
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```
# Agrupando estado com idcompra
vendas.groupby("estado").idcompra.nunique()
```

```
# Tirando os espaos ente as palavras e modificando para underline

vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```
# Chamando as informaes da tabela vendas
vendas.info()
```

```
# extraindo todos os valores nulos da colna estado

vendas[vendas['estado'].isnull()]
```

```
# Deletando os valores nulos da tebela
vendas_sem_nulos = vendas.dropna()

# verificando as informanes com novas na tabela sem os valores nulos
vendas_sem_nulos.info()
```

```
# Os valores nulos transformados para MS: Mato Grosso do Sul como solicitados.

# Sem inplace=True:
# Quando voc usa fillna() sem inplace=True (como no exemplo anterior),
# o Pandas cria uma cpia do DataFrame com os valores ausentes preenchidos.
# O DataFrame original vendas permanece inalterado. Para salvar as mudanas,
#  necessrio atribuir a cpia de volta ao DataFrame original, assim:
vendas["estado"] = vendas["estado"].fillna("MS")

# Com inplace=True:
# Quando voc usa fillna() com inplace=True, a modificao  feita diretamente no DataFrame original.
```

```
vendas['estado'].fillna("MS", inplace = True)

```

```
# vendas as informaes da tabela.
vendas.info()
```

```
# Media do Preo
media_preco = vendas["Preo"].mean()

vendas['Preo'].fillna(media_preco, inplace = True)
```

```
vendas.info()
```

```
# Uma Descrio da tabela Preo e Preo_com_frete
vendas[["Preo", "Preo_com_frete"]].describe()
```

```
# Retirando os preos que so maiores que o preo com frete
preco_errado = vendas.query("Preo > Preo_com_frete")
# Selecionando os preos que so menores que o preo com frete
vendas_correto = vendas.query("Preo < Preo_com_frete")
```

```
# Descrio da tabela vendas_correto
vendas_correto.describe()
```

```
# Filtrando a coluna  Nome_Departamento, das vendas Esporte e Lazer, e selecioando as vendas por estado de SP
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP' ").describe()
```

```
# Agrupando a vendas por Nome_Departamento com idcompra e calculado os valores unicos e classificando em uma ordem decrescente.
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```
# Deixando duas casa decimais, agrupando Nome_Departamento, tirando a media do Preo_com_frete, calculando os valores unicos
#classificando em ordem decrecente.
round(vendas_correto.groupby("Nome_Departamento")["Preo_com_frete"].agg("mean").sort_values(ascending = False).reset_index(),2)
```

```
# Agrupando a Data com o idcompra e calculas os valores unicos
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```
# Criando uma nova coluna Ms, para extrair o ms da coluna Data.
vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```
# Agrupando a coluna mes e calculando idcompra e colocando em ordem decrescente
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```
# Importar a tabela cliente_varejo do excel para python.
cliente = pd.read_excel("cliente_varejo.xlsx")
# monstrando a tabela.
cliente.head()
```

```
# Mostrando a tabela vendas_correto.
vendas_correto.head()
```

```
# Mostrando a descrio da tabela cliente e selecionando as colunas idade e renda
round(cliente[["idade", "renda"]].describe(),2)
```

```
# Mostando as Informao da tabela cliente
cliente.info()
```

```
# Transformando a coluna renda da tabela cliente para ponto flutuante.
cliente = cliente.astype({"renda": "float"})
cliente.info()
```

```
# Criando vareavel, chamando primeira tabela "vendas_correto", unindo as tabela com .merge, chama a segunda tabela "cliente" e define
# como quer ser feita a unio "how= 'left'", e chama a chave primaria das tabela "cliente_Log"
vendas_cliente = vendas_correto.merge(cliente, how = "left", on = "cliente_Log")
vendas_cliente.head()
```

```
# variavel = Duas casa decimais, agrupamento da coluna idcanalvenda, a media de renda, e a ordenao do maior para menor
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```
# variavel = Duas casas decimais, agrupamento da coluna bandeira e a media de idade, e a ordenao do maior para menor
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```
# Importando a biblioteca matplotlib.pyplot
import matplotlib.pyplot as plt
```

```
#Criando um container para o grfico usando a funo figure() do mdulo pyplot da biblioteca que foi importada com plt.
#figsize=(7,4): Este argumento define o tamanho da figura em polegadas, com 7 polegadas de largura e 4 polegadas de altura. Isso ajuda a controlar as dimenses do grfico.
fig = plt.figure(figsize=(7,4))

#plt.bar(): Esta linha  a que realmente cria o grfico de barras. A funo bar() do pyplot  usada para isso.
#agg_idade_bandeira["bandeira"]: Este argumento fornece os dados para o eixo horizontal (eixo x) do grfico.
#Presumivelmente, agg_idade_bandeira  um DataFrame (provavelmente do Pandas) e a coluna "bandeira" contm os nomes ou categorias das bandeiras.
#agg_idade_bandeira["idade"]: Este argumento fornece os dados para o eixo vertical (eixo y) do grfico. A coluna "idade" do DataFrame agg_idade_bandeira deve conter os valores da mdia de idade para cada bandeira.
#color="blue": Define a cor das barras do grfico como azul.
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="blue")

#Definindo o rtulo do eixo y como "Mdia de Idade".
plt.ylabel("Mdia de Idade")

#Esta linha define o ttulo do grfico como "Idade mdia por Bandeira".
#loc="center" centraliza o ttulo.
#fontsize=15 define o tamanho da fonte do ttulo como 15.
#fontweight="bold" define o estilo da fonte do ttulo como negrito.
#color="black" define a cor da fonte do ttulo como preta.
plt.title("Idade mdia por Bandeira", loc="center", fontsize=15, fontweight="bold", color="black");
```

```
#fig = plt.figure(...): Esta linha cria uma nova figura onde o grfico ser desenhado. A varivel fig armazena a referncia para essa figura.
#figsize=(7,4): Define o tamanho da figura em polegadas. Neste caso, a figura ter 7 polegadas de largura e 4 polegadas de altura.
fig = plt.figure(figsize=(7,4))

#plt.bar(...): Esta  a funo principal que cria o grfico de barras.
#agg_idcanal_renda["idcanalvenda"]: Este argumento fornece os dados para o eixo horizontal (eixo x) do grfico. Presumivelmente, agg_idcanal_renda  um DataFrame (provavelmente um Pandas DataFrame) e a coluna "idcanalvenda" contm os identificadores ou categorias dos canais de venda.
#agg_idcanal_renda["renda"]: Este argumento fornece os dados para o eixo vertical (eixo y) do grfico. A coluna "renda" do DataFrame provavelmente contm os valores da renda mdia correspondentes a cada canal de venda.
#color="blue": Define a cor das barras como azul.
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="blue")

#plt.ylabel(...): Adiciona um rtulo ao eixo y do grfico. Neste caso, o rtulo ser "Mdia de Renda".
plt.ylabel("Mdia de Renda")

#plt.title(...): Adiciona um ttulo ao grfico.
#"Renda Mdia por canal de venda": O texto do ttulo.
#loc="center": Posiciona o ttulo no centro do grfico.
#fontsize=15: Define o tamanho da fonte do ttulo para 15.
#fontweight="bold": Define o ttulo em negrito.
#color="black": Define a cor da fonte do ttulo para preto.
plt.title("Renda Mdia por canal de venda", loc="center", fontsize=15, fontweight="bold", color="black");
```

```
#vendas_correto.groupby("Data"): Esta parte agrupa os dados do DataFrame vendas_correto pela coluna "Data". Isso significa que todas as linhas com a mesma data sero agrupadas juntas.
#.idcompra.nunique(): Dentro de cada grupo (cada data), esta parte seleciona a coluna "idcompra" e aplica a funo nunique(). Esta funo conta o nmero de valores nicos dentro da coluna "idcompra". Em outras palavras, est contando o nmero de compras nicas para cada data.
#.reset_index(): Esta parte transforma o resultado do groupby (que  uma estrutura de ndice hierrquico) em um novo DataFrame. A coluna "Data" que era o ndice do groupby agora se torna uma coluna regular do novo DataFrame.
#venda_por_data = ...: O resultado de toda a operao (um novo DataFrame com a data e o nmero de vendas nicas por data)  atribudo  varivel venda_por_data.
#venda_por_data: Esta linha, por si s, exibe o contedo do DataFrame venda_por_data.

venda_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index();
venda_por_data
```

```
#Essa linha cria uma figura para o grfico usando a funo figure do Matplotlib.
#figsize=(20,10) define o tamanho da figura em polegadas, com 20 polegadas de largura e 10 polegadas de altura. Isso garante um grfico maior e mais fcil de visualizar.
fig = plt.figure(figsize=(20,10))

#Define o rtulo do eixo X como "Data da venda". Isso ajuda a identificar o que est sendo representado no eixo horizontal.
plt.xlabel("Data da venda")

#Define o rtulo do eixo Y como "Quantidade de vendas". Isso ajuda a identificar o que est sendo representado no eixo vertical.
plt.ylabel("Quantidade de vendas")

#Atribui os valores da coluna "Data" do DataFrame venda_por_data  varivel x. Esses valores sero usados como coordenadas no eixo X do grfico, representando as datas das vendas.
x = venda_por_data["Data"]

#Atribui os valores da coluna "idcompra" do DataFrame venda_por_data  varivel y. Esses valores sero usados como coordenadas no eixo Y do grfico, representando a quantidade de vendas para cada data.
y = venda_por_data["idcompra"]

#Cria o grfico de linha usando a funo plot do Matplotlib.
#x e y so passados como argumentos para definir as coordenadas dos pontos no grfico.
#color="blue" define a cor da linha do grfico como azul.
#linewidth = "2" define a espessura da linha do grfico como 2.
plt.plot(x,y, color="blue", linewidth = "2");

#Exibe o grfico gerado.
plt.show()
```

```
#Agregao por departamento e por preo.
#vendas_correto.groupby("Nome_Departamento"):Essa parte do cdigo agrupa os dados do DataFrame vendas_correto pela coluna "Nome_Departamento". Isso significa que os dados sero organizados em grupos, onde cada grupo representa um departamento diferente.
#["Preo_com_frete"]:Seleciona a coluna "Preo_com_frete" dentro de cada grupo de departamento. Isso indica que queremos analisar os preos com frete para cada departamento.
#.agg("mean"):Calcula a mdia dos valores da coluna "Preo_com_frete" para cada grupo de departamento. Ou seja, calcula o preo mdio com frete para cada departamento.
#.sort_values(ascending = False):Ordena os resultados em ordem decrescente, com base nos valores mdios de "Preo_com_frete". Isso significa que os departamentos com os preos mdios mais altos sero listados primeiro.
#ascending = False especifica que a ordenao deve ser feita em ordem decrescente.
#.reset_index(): Redefine o ndice do DataFrame resultante. Isso cria um novo ndice numrico sequencial para o DataFrame.
#round(...,2): Arredonda os valores da coluna "Preo_com_frete" para duas casas decimais. Isso torna os valores mais legveis e fceis de interpretar.
agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")["Preo_com_frete"].agg("mean").sort_values(ascending = False).reset_index(),2)

#agg_dept_preco = ...: Atribui o DataFrame resultante  varivel agg_dept_preco. Isso permite que voc acesse e utilize os dados posteriormente.
agg_dept_preco
```

```
#fig = plt.figure(figsize=(40,25)):Essa linha cria uma figura para o grfico de barras usando a funo figure do Matplotlib.
#figsize=(40,25): define o tamanho da figura em polegadas, resultando em um grfico grande com 40 polegadas de largura e 25 polegadas de altura, ideal para visualizar muitos departamentos.
fig = plt.figure(figsize=(40,25))

#plt.bar(...):Cria o grfico de barras usando a funo bar do Matplotlib.
#agg_dept_preco["Nome_Departamento"]: Define os rtulos do eixo X com os nomes dos departamentos, extrados da coluna "Nome_Departamento" do DataFrame agg_dept_preco.
#agg_dept_preco["Preo_com_frete"]: Define as alturas das barras com os preos mdios com frete, extrados da coluna "Preo_com_frete" do DataFrame agg_dept_preco.
#color="blue": Define a cor das barras como azul.
plt.bar(agg_dept_preco["Nome_Departamento"],  agg_dept_preco["Preo_com_frete"], color="blue")

#plt.ylabel("Preo mdio"):Define o rtulo do eixo Y como "Preo mdio", tornando claro que o eixo vertical representa o preo mdio com frete dos produtos.
plt.ylabel("Preo mdio")

#plt.title(...): Define o ttulo do grfico como "Preo mdio por departamento".
#loc="center": Centraliza o ttulo no topo do grfico.
#fontsize=15: Define o tamanho da fonte do ttulo como 15.
#fontweight="bold": Define o ttulo em negrito para destaque.
#color="black": Define a cor do ttulo como preto.
plt.title("Preo mdio por departamento", loc="center", fontsize=15, fontweight="bold", color="black");
```
