```
import pandas as pd
import seaborn as sns
```

```
# Importando a base de dados da tabela excel
vendas = pd.read_excel('/content/varejo.xlsx')
```

```
# Chamando a tabela
vendas.head()
```

```
# Chamando as informações da tabela para analizar os valores nulos e para ver os dados de cada coluna. Tipo: str,object,float,int.
vendas.info()
```

```
# Agrupando Idcanalvenda com idcompra.
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```
# Modificando a abreviação de APP para todos os dados ficar iguais como Aplicativo
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```
# Agrupando as bandeira com idcompra
vendas.groupby("bandeira").idcompra.nunique()
```

```
# Agrupando Nome_Departamento com idcompra
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```
# Agrupando estado com idcompra
vendas.groupby("estado").idcompra.nunique()
```

```
# Tirando os espaços ente as palavras e modificando para underline

vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```
# Chamando as informações da tabela vendas
vendas.info()
```

```
# extraindo todos os valores nulos da colna estado

vendas[vendas['estado'].isnull()]
```

```
# Deletando os valores nulos da tebela
vendas_sem_nulos = vendas.dropna()

# verificando as informanções com novas na tabela sem os valores nulos
vendas_sem_nulos.info()
```

```
# Os valores nulos transformados para MS: Mato Grosso do Sul como solicitados.

# Sem inplace=True:
# Quando você usa fillna() sem inplace=True (como no exemplo anterior),
# o Pandas cria uma cópia do DataFrame com os valores ausentes preenchidos.
# O DataFrame original vendas permanece inalterado. Para salvar as mudanças,
# é necessário atribuir a cópia de volta ao DataFrame original, assim:
vendas["estado"] = vendas["estado"].fillna("MS")

# Com inplace=True:
# Quando você usa fillna() com inplace=True, a modificação é feita diretamente no DataFrame original.
```

```
vendas['estado'].fillna("MS", inplace = True)

```

```
# vendas as informações da tabela.
vendas.info()
```

```
# Media do Preço
media_preco = vendas["Preço"].mean()

vendas['Preço'].fillna(media_preco, inplace = True)
```

```
vendas.info()
```

```
# Uma Descrição da tabela Preço e Preço_com_frete
vendas[["Preço", "Preço_com_frete"]].describe()
```

```
# Retirando os preços que são maiores que o preço com frete
preco_errado = vendas.query("Preço > Preço_com_frete")
# Selecionando os preços que são menores que o preço com frete
vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```
# Descrição da tabela vendas_correto
vendas_correto.describe()
```

```
# Filtrando a coluna  Nome_Departamento, das vendas Esporte e Lazer, e selecioando as vendas por estado de SP
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP' ").describe()
```

```
# Agrupando a vendas por Nome_Departamento com idcompra e calculado os valores unicos e classificando em uma ordem decrescente.
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```
# Deixando duas casa decimais, agrupando Nome_Departamento, tirando a media do Preço_com_frete, calculando os valores unicos
#classificando em ordem decrecente.
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending = False).reset_index(),2)
```

```
# Agrupando a Data com o idcompra e calculas os valores unicos
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```
# Criando uma nova coluna Mês, para extrair o mês da coluna Data.
vendas_correto["mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```
# Agrupando a coluna mes e calculando idcompra e colocando em ordem decrescente
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending = False).reset_index()
```

```
# Importar a tabela cliente_varejo do excel para python.
cliente = pd.read_excel("cliente_varejo.xlsx")
# monstrando a tabela.
cliente.head()
```

```
# Mostrando a tabela vendas_correto.
vendas_correto.head()
```

```
# Mostrando a descrição da tabela cliente e selecionando as colunas idade e renda
round(cliente[["idade", "renda"]].describe(),2)
```

```
# Mostando as Informação da tabela cliente
cliente.info()
```

```
# Transformando a coluna renda da tabela cliente para ponto flutuante.
cliente = cliente.astype({"renda": "float"})
cliente.info()
```

```
# Criando vareavel, chamando primeira tabela "vendas_correto", unindo as tabela com .merge, chama a segunda tabela "cliente" e define
# como quer ser feita a união "how= 'left'", e chama a chave primaria das tabela "cliente_Log"
vendas_cliente = vendas_correto.merge(cliente, how = "left", on = "cliente_Log")
vendas_cliente.head()
```

```
# variavel = Duas casa decimais, agrupamento da coluna idcanalvenda, a media de renda, e a ordenação do maior para menor
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```
# variavel = Duas casas decimais, agrupamento da coluna bandeira e a media de idade, e a ordenação do maior para menor
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```
# Importando a biblioteca matplotlib.pyplot
import matplotlib.pyplot as plt
```

```
#Criando um container para o gráfico usando a função figure() do módulo pyplot da biblioteca que foi importada com plt.
#figsize=(7,4): Este argumento define o tamanho da figura em polegadas, com 7 polegadas de largura e 4 polegadas de altura. Isso ajuda a controlar as dimensões do gráfico.
fig = plt.figure(figsize=(7,4))

#plt.bar(): Esta linha é a que realmente cria o gráfico de barras. A função bar() do pyplot é usada para isso.
#agg_idade_bandeira["bandeira"]: Este argumento fornece os dados para o eixo horizontal (eixo x) do gráfico.
#Presumivelmente, agg_idade_bandeira é um DataFrame (provavelmente do Pandas) e a coluna "bandeira" contém os nomes ou categorias das bandeiras.
#agg_idade_bandeira["idade"]: Este argumento fornece os dados para o eixo vertical (eixo y) do gráfico. A coluna "idade" do DataFrame agg_idade_bandeira deve conter os valores da média de idade para cada bandeira.
#color="blue": Define a cor das barras do gráfico como azul.
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="blue")

#Definindo o rótulo do eixo y como "Média de Idade".
plt.ylabel("Média de Idade")

#Esta linha define o título do gráfico como "Idade média por Bandeira".
#loc="center" centraliza o título.
#fontsize=15 define o tamanho da fonte do título como 15.
#fontweight="bold" define o estilo da fonte do título como negrito.
#color="black" define a cor da fonte do título como preta.
plt.title("Idade média por Bandeira", loc="center", fontsize=15, fontweight="bold", color="black");
```

```
#fig = plt.figure(...): Esta linha cria uma nova figura onde o gráfico será desenhado. A variável fig armazena a referência para essa figura.
#figsize=(7,4): Define o tamanho da figura em polegadas. Neste caso, a figura terá 7 polegadas de largura e 4 polegadas de altura.
fig = plt.figure(figsize=(7,4))

#plt.bar(...): Esta é a função principal que cria o gráfico de barras.
#agg_idcanal_renda["idcanalvenda"]: Este argumento fornece os dados para o eixo horizontal (eixo x) do gráfico. Presumivelmente, agg_idcanal_renda é um DataFrame (provavelmente um Pandas DataFrame) e a coluna "idcanalvenda" contém os identificadores ou categorias dos canais de venda.
#agg_idcanal_renda["renda"]: Este argumento fornece os dados para o eixo vertical (eixo y) do gráfico. A coluna "renda" do DataFrame provavelmente contém os valores da renda média correspondentes a cada canal de venda.
#color="blue": Define a cor das barras como azul.
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="blue")

#plt.ylabel(...): Adiciona um rótulo ao eixo y do gráfico. Neste caso, o rótulo será "Média de Renda".
plt.ylabel("Média de Renda")

#plt.title(...): Adiciona um título ao gráfico.
#"Renda Média por canal de venda": O texto do título.
#loc="center": Posiciona o título no centro do gráfico.
#fontsize=15: Define o tamanho da fonte do título para 15.
#fontweight="bold": Define o título em negrito.
#color="black": Define a cor da fonte do título para preto.
plt.title("Renda Média por canal de venda", loc="center", fontsize=15, fontweight="bold", color="black");
```

```
#vendas_correto.groupby("Data"): Esta parte agrupa os dados do DataFrame vendas_correto pela coluna "Data". Isso significa que todas as linhas com a mesma data serão agrupadas juntas.
#.idcompra.nunique(): Dentro de cada grupo (cada data), esta parte seleciona a coluna "idcompra" e aplica a função nunique(). Esta função conta o número de valores únicos dentro da coluna "idcompra". Em outras palavras, está contando o número de compras únicas para cada data.
#.reset_index(): Esta parte transforma o resultado do groupby (que é uma estrutura de índice hierárquico) em um novo DataFrame. A coluna "Data" que era o índice do groupby agora se torna uma coluna regular do novo DataFrame.
#venda_por_data = ...: O resultado de toda a operação (um novo DataFrame com a data e o número de vendas únicas por data) é atribuído à variável venda_por_data.
#venda_por_data: Esta linha, por si só, exibe o conteúdo do DataFrame venda_por_data.

venda_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index();
venda_por_data
```

```
#Essa linha cria uma figura para o gráfico usando a função figure do Matplotlib.
#figsize=(20,10) define o tamanho da figura em polegadas, com 20 polegadas de largura e 10 polegadas de altura. Isso garante um gráfico maior e mais fácil de visualizar.
fig = plt.figure(figsize=(20,10))

#Define o rótulo do eixo X como "Data da venda". Isso ajuda a identificar o que está sendo representado no eixo horizontal.
plt.xlabel("Data da venda")

#Define o rótulo do eixo Y como "Quantidade de vendas". Isso ajuda a identificar o que está sendo representado no eixo vertical.
plt.ylabel("Quantidade de vendas")

#Atribui os valores da coluna "Data" do DataFrame venda_por_data à variável x. Esses valores serão usados como coordenadas no eixo X do gráfico, representando as datas das vendas.
x = venda_por_data["Data"]

#Atribui os valores da coluna "idcompra" do DataFrame venda_por_data à variável y. Esses valores serão usados como coordenadas no eixo Y do gráfico, representando a quantidade de vendas para cada data.
y = venda_por_data["idcompra"]

#Cria o gráfico de linha usando a função plot do Matplotlib.
#x e y são passados como argumentos para definir as coordenadas dos pontos no gráfico.
#color="blue" define a cor da linha do gráfico como azul.
#linewidth = "2" define a espessura da linha do gráfico como 2.
plt.plot(x,y, color="blue", linewidth = "2");

#Exibe o gráfico gerado.
plt.show()
```

```
#Agregação por departamento e por preço.
#vendas_correto.groupby("Nome_Departamento"):Essa parte do código agrupa os dados do DataFrame vendas_correto pela coluna "Nome_Departamento". Isso significa que os dados serão organizados em grupos, onde cada grupo representa um departamento diferente.
#["Preço_com_frete"]:Seleciona a coluna "Preço_com_frete" dentro de cada grupo de departamento. Isso indica que queremos analisar os preços com frete para cada departamento.
#.agg("mean"):Calcula a média dos valores da coluna "Preço_com_frete" para cada grupo de departamento. Ou seja, calcula o preço médio com frete para cada departamento.
#.sort_values(ascending = False):Ordena os resultados em ordem decrescente, com base nos valores médios de "Preço_com_frete". Isso significa que os departamentos com os preços médios mais altos serão listados primeiro.
#ascending = False especifica que a ordenação deve ser feita em ordem decrescente.
#.reset_index(): Redefine o índice do DataFrame resultante. Isso cria um novo índice numérico sequencial para o DataFrame.
#round(...,2): Arredonda os valores da coluna "Preço_com_frete" para duas casas decimais. Isso torna os valores mais legíveis e fáceis de interpretar.
agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending = False).reset_index(),2)

#agg_dept_preco = ...: Atribui o DataFrame resultante à variável agg_dept_preco. Isso permite que você acesse e utilize os dados posteriormente.
agg_dept_preco
```

```
#fig = plt.figure(figsize=(40,25)):Essa linha cria uma figura para o gráfico de barras usando a função figure do Matplotlib.
#figsize=(40,25): define o tamanho da figura em polegadas, resultando em um gráfico grande com 40 polegadas de largura e 25 polegadas de altura, ideal para visualizar muitos departamentos.
fig = plt.figure(figsize=(40,25))

#plt.bar(...):Cria o gráfico de barras usando a função bar do Matplotlib.
#agg_dept_preco["Nome_Departamento"]: Define os rótulos do eixo X com os nomes dos departamentos, extraídos da coluna "Nome_Departamento" do DataFrame agg_dept_preco.
#agg_dept_preco["Preço_com_frete"]: Define as alturas das barras com os preços médios com frete, extraídos da coluna "Preço_com_frete" do DataFrame agg_dept_preco.
#color="blue": Define a cor das barras como azul.
plt.bar(agg_dept_preco["Nome_Departamento"],  agg_dept_preco["Preço_com_frete"], color="blue")

#plt.ylabel("Preço médio"):Define o rótulo do eixo Y como "Preço médio", tornando claro que o eixo vertical representa o preço médio com frete dos produtos.
plt.ylabel("Preço médio")

#plt.title(...): Define o título do gráfico como "Preço médio por departamento".
#loc="center": Centraliza o título no topo do gráfico.
#fontsize=15: Define o tamanho da fonte do título como 15.
#fontweight="bold": Define o título em negrito para destaque.
#color="black": Define a cor do título como preto.
plt.title("Preço médio por departamento", loc="center", fontsize=15, fontweight="bold", color="black");
```
